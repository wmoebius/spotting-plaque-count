```python
import matplotlib.pyplot as plt
from phageid.processing.layers import GaussianBlur, SubtractByFrame, MaskReplace, Convolution, PeakFinder, Difference
from phageid.processing.kernels import GaussianKernel, CircularKernel
from phageid.utils import find_first_peak
```

```python
# sample = "sample_04_05.npy"
# sample = "sample_04_06.npy"
sample = "sample_02_06.npy"
# sample = "sample_03_08.npy"

file_path = Path(f"/home/finley/Work/RDS/projects/phageid/data/processed/splits_2025-01-31_15-27-52/{sample}")
assert file_path.is_file()
data = np.load(file_path).astype(float)
data_ = data.copy()
plt.imshow(data[12])
```

```python
# mask out debris
mask = data[0]<180
```

```python
# normalise to background intensity
data = SubtractByFrame(value=lambda x: find_first_peak(x, 256, False))(data)
plt.imshow(data[12])
```

```python
data = GaussianBlur(kernel_size=7, sigma=.5)(data)
plt.imshow(data[12])
```

```python
kernel = GaussianKernel(size=13, sigma=2)
# kernel._kernel -= kernel._kernel.mean()
plt.imshow(kernel._kernel)
plt.colorbar()
```

```python
detected = Convolution(kernel=kernel)(data)
plt.imshow(detected[12])
```

```python
detected = GaussianBlur(kernel_size=9, sigma=1)(detected)
plt.imshow(detected[12])
```

```python
peaks = PeakFinder(min_distance = 15,
threshold_rel= 0.0,
threshold_abs= 32)(detected)
# 105
```

```python
video_image_sequence_with_scatter(detected, peaks, fps=10)
```
