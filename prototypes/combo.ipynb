```python
from phageid.detection.detection import RingDetector, GaussianDetector
from phageid.visualisation import video_image_sequence, video_image_sequence_with_scatter
from phageid.processing.layers import AgglomeratePeaks
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# load data
sample = "sample_04_05.npy"
# sample = "sample_04_06.npy"
sample = "sample_02_06.npy"
# sample = "sample_03_08.npy"

file_path = Path(f"/home/finley/Work/RDS/projects/phageid/data/processed/splits_2025-01-31_15-27-52/{sample}")
data = np.load(file_path)
data_ = data.copy()
```

```python
video_image_sequence(data)
```

```python
points_ring = RingDetector()(data_.copy())
points_gaus = GaussianDetector()(data_.copy())
```

```python
combined = [np.vstack((r, g)) for r, g in zip(points_ring, points_gaus)]
```

```python
video_image_sequence_with_scatter(data[2:], combined)
```

```python
plt.imshow(data[-1])
plt.scatter(*np.vstack(combined).T, c="r", s=5)
```

```python
agglommed = AgglomeratePeaks(min_distance=15)(combined)
plt.imshow(data[-1])
plt.scatter(*np.vstack(agglommed).T, c="r", s=5)
```
